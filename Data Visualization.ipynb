Matplot lib

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

x = np.linspace(5,100,100)
y = np.linspace(10,1000,100)

plt.xlabel("Time")
plt.ylabel("Price")
plt.title("Market Price")

plt.xlim([20,80])
plt.ylim([200,500])
plt.plot(x,y)
plt.show()
```

```python
x = np.linspace(5,100,100)
y = np.linspace(10,1000,100)

plt.xlabel("Time")
plt.ylabel("Price")
plt.title("Market Price")

plt.xlim([20,80])
plt.ylim([200,500])
plt.plot(x,y, 'b+')   # to print this blue +
plt.show()
```

# Figures and Subplots
You often need to create multiple plots in the same figure, as we'll see in some upcoming examples.

matplotlib has the concept of figures and subplots using which you can create multiple subplots inside the same figure.

To create multiple plots in the same figure, you can use the method plt.subplot(nrows, ncols, nsubplot).

```python
#Subplot


x2 = np.linspace(1,20,10)

plt.figure(1)
plt.subplot(2,2,1)
plt.plot(x2,x2**2)
plt.title("Squre")

plt.subplot(2,2,2)
plt.plot(x2,x2**3)
plt.title("Qube")

plt.figure(2)
plt.subplot(2,2,1)
plt.plot(x2,np.log(x2))
plt.title("Log")


plt.subplot(2,2,2)
plt.plot(x2,x2)
plt.title("Linear")

plt.show()
```
