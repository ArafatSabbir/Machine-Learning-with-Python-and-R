```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

x = np.linspace(5,100,100)
y = np.linspace(10,1000,100)

plt.xlabel("Time")
plt.ylabel("Price")
plt.title("Market Price")

plt.xlim([20,80])
plt.ylim([200,500])
plt.plot(x,y)
```
